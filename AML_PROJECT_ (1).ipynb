````python
import matplotlib.pyplot as plt
from kmodes.kprototypes import KPrototypes
from tqdm import tqdm # Import tqdm for progress bar
import multiprocessing as mp # Import multiprocessing


#1. Prepare the data Select categorical and numerical features
categorical_cols = banking_data.select_dtypes(include=['object', 'category']).columns.tolist()
numerical_cols = banking_data.select_dtypes(include=np.number).columns.tolist()

# Combine categorical and numerical features
all_cols = categorical_cols + numerical_cols
data_for_clustering = banking_data[all_cols]

# 2. Apply Elbow Method with k-prototypes (with progress bar)

def calculate_cost(num_clusters): # Function to calculate cost for a single cluster number
    kproto = KPrototypes(n_clusters=num_clusters, init='Cao', verbose=0) # Set verbose=0 to suppress output
    clusters = kproto.fit_predict(data_for_clustering, categorical=list(range(len(categorical_cols))))
    return kproto.cost_

cluster_range = list(range(1, 15))
with mp.Pool(processes=mp.cpu_count()) as pool: # Use multiprocessing pool
    cost = list(tqdm(pool.imap(calculate_cost, cluster_range), total=len(cluster_range), desc="Calculating costs"))

# 3. Visualize the Elbow Method results
plt.plot(cluster_range, cost, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Cost')
plt.title('Elbow Method for Optimal k (Mixed Data)')
plt.show()
````

## Aggloromative Clustering

````python
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.preprocessing import StandardScaler, OneHotEncoder
import pandas as pd


````

````python


# Assuming 'banking_data' is your DataFrame

# 1. Select relevant features for PCA
numeric_columns = ['age', 'transaction_frequency', 'total_tx_volume',
                   'avg_tx_value', 'satisfaction_score', 'nps_score',
                   'active_products', 'churn_probability', 'customer_lifetime_value',
                   'total_transaction_volume', 'monthly_transaction_count', 'avg_daily_transactions']
categorical_columns = ['income_bracket', 'occupation', 'customer_segment', 'education_level',
                       'marital_status', 'acquisition_channel', 'savings_account',
                       'credit_card', 'personal_loan', 'investment_account']

# 2. Preprocess the data
scaler = StandardScaler()
scaled_numeric_data = scaler.fit_transform(banking_data[numeric_columns])

encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')  # Handle unknown categories
encoded_categorical_data = encoder.fit_transform(banking_data[categorical_columns])

# 3. Combine scaled numeric and encoded categorical features
combined_data = pd.concat([pd.DataFrame(scaled_numeric_data, columns=numeric_columns),
                           pd.DataFrame(encoded_categorical_data,
                                        columns=encoder.get_feature_names_out(categorical_columns))], axis=1)

# we use 'combined_data' in the linkage function
Z = linkage(combined_data, method='ward')
````

````python
plt.figure(figsize=(10, 7))
dendrogram(Z, orientation='top', labels=banking_data['customer_id'].tolist(),
           distance_sort='descending', show_leaf_counts=True)
plt.title('Dendrogram of Agglomerative Clustering')
plt.xlabel('Customer ID')
plt.ylabel('Distance')
plt.show()
````

MAKING PCA_3 AND PCA_4 TO GENERATE CUSTOMER CHRUN PREDICTION AND HENCE LATER IDENTIFYING THE LIKEHOOD Of the customer staying

````python

from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix, mean_squared_error, r2_score

````

````python
# Load dataset
banking_data = pd.read_excel("/content/banking_data.xlsx", engine='openpyxl')

# Inspect the dataset
print(banking_data.head())
print(banking_data.info())

# Define features and target
features = ['satisfaction_score', 'nps_score', 'savings_account',
            'credit_card', 'personal_loan', 'investment_account',
            'income_bracket', 'occupation', 'customer_segment', 'education_level',
            'marital_status', 'acquisition_channel']
target = 'churn_probability'
````

````python
# Filter the first 100 rows
subset_data = banking_data.iloc[:100]

# Separate features and target
X = subset_data[features]
y = subset_data[target]

# Standardize the numerical features
scaler = StandardScaler()
numerical_features = ['satisfaction_score', 'nps_score', 'savings_account',
                      'credit_card', 'personal_loan', 'investment_account']
scaled_features = scaler.fit_transform(subset_data[numerical_features])

# One-hot encode categorical features
categorical_columns = ['income_bracket', 'occupation', 'customer_segment',
                       'education_level', 'marital_status', 'acquisition_channel']
encoder = OneHotEncoder()
encoded_features = encoder.fit_transform(subset_data[categorical_columns]).toarray()

# Combine scaled numerical and encoded categorical features
X_combined = pd.concat([pd.DataFrame(scaled_features, columns=numerical_features),
                        pd.DataFrame(encoded_features, columns=encoder.get_feature_names_out(categorical_columns))], axis=1)
````

````python

pca = PCA(n_components=2)
pca_features = pca.fit_transform(X_combined)
subset_data['pca_3'] = pca_features[:, 0]
subset_data['pca_4'] = pca_features[:, 1]

````

````python
# Apply K-Means clustering
n_clusters = 2
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
subset_data['cluster'] = kmeans.fit_predict(subset_data[['pca_3', 'pca_4']])

# Visualize Clusters
colors = ['red', 'blue']
cluster_colors = [colors[label] for label in subset_data['cluster']]
plt.scatter(subset_data['pca_3'], subset_data['pca_4'], c=cluster_colors, alpha=0.7)
plt.title('K-Means Clustering on PCA Features')
plt.xlabel('PCA 3')
plt.ylabel('PCA 4')
plt.show()
````

````python
#Prepare data for Random Forest
from sklearn.metrics import mean_squared_error, r2_score # Import mean_squared_error and r2_score

X_clustered = subset_data[['pca_3', 'pca_4', 'cluster']]
X_train, X_test, y_train, y_test = train_test_split(X_clustered, y, test_size=0.2, random_state=42)

# Train Random Forest Regressor
rf_model = RandomForestRegressor(random_state=42)
rf_model.fit(X_train, y_train)

# Evaluate the model
y_pred = rf_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Mean Squared Error: {mse:.4f}")
print(f"R-squared: {r2:.4f}")
````

````python
# Confusion Matrix for Classification-Like Analysis (Threshold Based)
threshold = 0.5
y_pred_class = [1 if prob >= threshold else 0 for prob in y_pred]
y_test_class = [1 if prob >= threshold else 0 for prob in y_test]
conf_matrix = confusion_matrix(y_test_class, y_pred_class)
print("Confusion Matrix:")
print(conf_matrix)

# Calculate stay likelihood
subset_data['stay_likelihood'] = (1 - subset_data['churn_probability']) * 100
print(subset_data[['customer_id', 'stay_likelihood']].head())

# Visualize Clusters vs Churn Probability
plt.scatter(subset_data['pca_3'], subset_data['pca_4'], c=subset_data['churn_probability'], cmap='coolwarm', alpha=0.7)
plt.title("Clusters vs Churn Probability")
plt.xlabel("PCA 3")
plt.ylabel("PCA 4")
plt.colorbar(label="Churn Probability")
plt.show()
````
